##### Objective 1 - Collect the salary spent for the world series champions by year


```python
# resetting index of salary_df so 'yearID', 'teamID' and 'lgID' become columns instead of indexes
salary_year_df = salary_year_df.reset_index()
salary_year_df.head()
```

```python
# Renameing column of the ws_champs_df to match salary_year_df so I can then merge the two Data Frames using the columns IDs
ws_champs_df = ws_champs_df.rename(columns={'teamIDwinner':'teamID'})
ws_champs_df.head()
```

```python
# Merging ws_champs_df with salary_year_df to calculate yearly salary spent for the champs
champs_yearly_salary_df = pd.merge(ws_champs_df, salary_year_df, on=['yearID','teamID'], how='left')
champs_yearly_salary_df.head()
```

##### Objective 2 - Collect the average salary spent for the losers by year

```python
# removing champs from the salary_year Data Frame
losers_salary_df = pd.merge(salary_year_df,ws_champs_df, on=['yearID','teamID'], how='left')
losers_salary_df = losers_salary_df[losers_salary_df['round'].isnull()]
losers_salary_df
```

```python
losers_ave_yearly_salary = losers_salary_df.groupby(['yearID'])[['salary']].mean().reset_index()
losers_ave_yearly_salary
```
