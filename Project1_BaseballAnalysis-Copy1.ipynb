```python
#Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
#Read in Teams CSV Data as DataFrame
teams_filepath = "Resources/Teams.csv"
teams_df = pd.read_csv(teams_filepath)
```

```python
#Organize Teams Dataframe Columns (Put Columns in Order, Get Rid of Columns you dont need)
teams_df = teams_df[["name", "yearID", "G", "W", "L", "R", "AB", "H", "2B", "3B", "HR", "BB", "SO", "SB", "ER", "ERA", "CG", "E"]]
```

```python
# Abbreviation Key

# G: Games, W: Wins, L: Losses, AB: At Bats, H: Hits, 2B: Doubles, 3B: Triples
# HR: Home Runs, BB: Walks, SO: Strikeouts, SB: Stolen Bases, ER: Earned Runs,
# ERA: Earned Run Average, CG: Complete Games, E: Errors
```

```python
#Drop all years prior to 1960 (Just analyzing the modern era of baseball ~1960)
teams_df = teams_df.loc[teams_df["yearID"] > 1959]
teams_df.head()
```

```python
# KM - Use only years from 1985 to 2015 (Our analysis is 1985<=2015)
start_year= 1985
end_year= 2015
```

```python
# KM - Use only years from 1985 to 2015 (Our analysis is 1985<=2015)
years_df= (teams_df['yearID'] >= start_year) & (teams_df['yearID'] <= end_year)
years_df.head()
```

```python
# KM - Use only years from 1985 to 2015 (Our analysis is 1985<=2015)
teams_df= teams_df.loc[years_df]
teams_df.head()
```
